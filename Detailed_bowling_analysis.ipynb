# **1. Overall Bowling Performance**

objectives 
- Most wickets taken in IPL history
- Best bowling averages (min 50 overs bowled)
-  Best economy rates (min 50 overs bowled)

```python
# Step 1: Count total wickets (excluding run-outs)
bowler_wickets = (deliveries[deliveries['is_wicket'] == 1]
                   .query("dismissal_kind != 'run out'")
                   .groupby('bowler')['is_wicket']
                   .count()
                   .reset_index()
                   .rename(columns={'is_wicket': 'total_wickets'}))
bowler_wickets
```

```python
# Step 2: Calculate total runs conceded & balls bowled (ignoring extras)
bowling_stats = (deliveries.groupby('bowler')
                 .agg(total_runs_conceded=('total_runs', 'sum'),
                      balls_bowled=('ball', 'count'))
                 .reset_index())
bowling_stats
```

```python
# Merge with wicket data
bowling_stats = bowling_stats.merge(bowler_wickets, on='bowler', how='left').fillna(0)
bowling_stats
```

```python
# Filter bowlers who bowled at least 300 balls
qualified_bowlers = bowling_stats[bowling_stats['balls_bowled'] >= 300].copy()
qualified_bowlers
```

```python
# Step 3: Calculate Bowling Average & Economy Rate
qualified_bowlers['bowling_average'] = qualified_bowlers['total_runs_conceded'] / qualified_bowlers['total_wickets']
qualified_bowlers['economy_rate'] = (qualified_bowlers['total_runs_conceded'] / qualified_bowlers['balls_bowled']) * 6
qualified_bowlers
```

```python
# Handling cases where bowler has no wickets (to avoid division by zero)
qualified_bowlers.replace({'bowling_average': {float('inf'): None}}, inplace=True)
```

```python
# Sorting results
most_wickets = qualified_bowlers.sort_values(by='total_wickets', ascending=False)
best_bowling_average = qualified_bowlers.sort_values(by='bowling_average', ascending=True)
best_economy_rate = qualified_bowlers.sort_values(by='economy_rate', ascending=True)
```

# **2.Match-Winning Bowling Performances Analysis**

```python
# Step 1: Count 3-Wicket & 5-Wicket Hauls
bowling_performance = (deliveries[deliveries['is_wicket'] == 1]
                        .groupby(['match_id', 'bowler'])
                        .agg(wickets=('is_wicket', 'sum'),
                             runs_conceded=('total_runs', 'sum'),
                             balls_bowled=('ball', 'count'))
                        .reset_index())
```

```python
bowling_performance
```

```python
# Step 2: Calculate Economy Rate (Runs Conceded per Over)
bowling_performance['economy_rate'] = (bowling_performance['runs_conceded'] / 
                                       (bowling_performance['balls_bowled'] / 6))
bowling_performance
```

```python
# Step 3: Identify 3-Wicket & 5-Wicket Hauls
three_wicket_hauls = bowling_performance[bowling_performance['wickets'] >= 3]
five_wicket_hauls = bowling_performance[bowling_performance['wickets'] >= 5]
```

```python
three_wicket_hauls
```

```python
five_wicket_hauls
```

```python

# Step 4: Count Match-Winning Performances by Bowler
match_winning_spells = (three_wicket_hauls.groupby('bowler')
                        .agg(three_wicket_hauls=('wickets', 'count'))
                        .reset_index())
```

```python
five_wicket_counts = (five_wicket_hauls.groupby('bowler')
                      .agg(five_wicket_hauls=('wickets', 'count'))
                      .reset_index())
```

```python
# Merge the two dataframes
match_winning_spells = match_winning_spells.merge(five_wicket_counts, on='bowler', how='left').fillna(0)
```

```python

# Sort by most impactful spells
match_winning_spells = match_winning_spells.sort_values(by=['three_wicket_hauls', 'five_wicket_hauls'], ascending=False)
match_winning_spells
```
