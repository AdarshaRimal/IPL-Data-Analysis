```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading datset
matches = pd.read_csv('matches.csv')
deleveries = pd.read_csv('deliveries.csv')
```

```python
matches.head()
```

# **Team Performance Over Seasons**

```python
team_matches = matches.melt(id_vars=["season"], value_vars=["team1", "team2"], 
                            var_name="team_column", value_name="team")
```

```python
team_matches
```

```python
# Count matches played by each team per season
matches_played = team_matches.groupby(["season", "team"]).size().reset_index(name="matches_played")
```

```python
matches_played
```

```python
# Count matches won by each team per season
matches_won = matches.groupby(["season", "winner"]).size().reset_index(name="matches_won")
```

```python
matches_won
```

```python
matches_won.rename(columns={"winner": "team"}, inplace=True)
```

```python
matches_won
```

```python
# Merge matches_played and matches_won on 'season' and 'team'
team_performance = pd.merge(matches_played, matches_won, on=["season", "team"], how="left")
team_performance
```

```python
# Calculate win percentage
team_performance['win_percentage'] = (team_performance['matches_won'] / team_performance['matches_played']) * 100
team_performance
```

```python
# Plotting win percentage over seasons for each team
plt.figure(figsize=(12, 6))
for team in team_performance['team'].unique():
    team_data = team_performance[team_performance['team'] == team]
    plt.plot(team_data['season'], team_data['win_percentage'], label=team)

# Adding labels and title
plt.title("Win Percentage Trend Over Seasons for Each Team", fontsize=14)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Win Percentage", fontsize=12)
plt.xticks(rotation=45)
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1))
plt.tight_layout()
plt.show()
```

```python
# Plotting a bar chart for top 5 teams in each season based on win percentage
top_teams = team_performance.groupby(['season', 'team']).agg({'win_percentage': 'max'}).reset_index()

plt.figure(figsize=(12, 6))
top_teams_sorted = top_teams.sort_values(by=['season', 'win_percentage'], ascending=[True, False])
for season in top_teams_sorted['season'].unique():
    season_data = top_teams_sorted[top_teams_sorted['season'] == season]
    plt.bar(season_data['team'], season_data['win_percentage'], label=season)

# Adding labels and title
plt.title("Top Teams by Win Percentage Per Season", fontsize=14)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Win Percentage", fontsize=12)
plt.xticks(rotation=-70)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivot table to plot heatmap
team_season_performance = team_performance.pivot(index="team", columns="season", values="win_percentage")

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(team_season_performance, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=.5)
plt.title("Team Performance Across Seasons (Win Percentage)", fontsize=14)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Team", fontsize=12)
plt.tight_layout()
plt.show()
```

# **Toss Impact on Winning**

```python
# Creating a new DataFrame to compare toss winners and match winners
toss_match_comparison = matches[['season', 'toss_winner', 'winner', 'toss_decision']]
toss_match_comparison['toss_win'] = toss_match_comparison['toss_winner'] == toss_match_comparison['winner']
toss_match_comparison
```

```python
# Separate data based on toss decision (bat-first or bowl-first)
bat_first = toss_match_comparison[toss_match_comparison['toss_decision'] == 'bat']
bowl_first = toss_match_comparison[toss_match_comparison['toss_decision'] == 'field']
bat_first
```

```python
bowl_first
```

```python
# Calculate win percentage for toss winners in both scenarios
bat_first_win_percentage = bat_first['toss_win'].mean() * 100  # Percentage of toss winners who won after batting first
bowl_first_win_percentage = bowl_first['toss_win'].mean() * 100  # Percentage of toss winners who won after bowling first
bat_first_win_percentage
```

```python
bowl_first_win_percentage
```

```python
# Prepare data for plotting
win_percentage_data = {
    "Bat First": bat_first_win_percentage,
    "Bowl First": bowl_first_win_percentage
}

# Plotting the bar chart
plt.figure(figsize=(8, 5))
plt.bar(win_percentage_data.keys(), win_percentage_data.values(), color=['orange', 'blue'])

# Adding labels and title
plt.title("Toss Impact on Winning (Bat First vs Bowl First)", fontsize=14)
plt.ylabel("Win Percentage", fontsize=12)
plt.xlabel("Toss Decision", fontsize=12)
plt.ylim(0, 100)  # Limiting y-axis to 0-100 for better visualization
plt.tight_layout()
plt.show()
```

# **Venue-Based Performance**

```python
# Group by venue and team to count matches played and won
venue_team_performance = matches.groupby(['venue', 'team1'])['winner'].value_counts().unstack(fill_value=0)
venue_team_performance
```

```python
# Count the total number of matches played (including both team1 and team2)
venue_team_performance['matches_played'] = venue_team_performance.sum(axis=1)
venue_team_performance
```

```python
# To count matches won, look for the value_counts of the 'winner' column, which will give you wins by the teams.
# We want to get the count of wins for each team at each venue.
venue_team_performance['matches_won'] = venue_team_performance[venue_team_performance.columns[0]]
venue_team_performance
```

```python
# Calculate the win percentage
venue_team_performance['win_percentage'] = (venue_team_performance['matches_won'] / venue_team_performance['matches_played']) * 100
venue_team_performance
```

```python
venue_team_performance = venue_team_performance.reset_index()
venue_team_performance
```

```python
# Create the pivot table
venue_team_performance_pivot = venue_team_performance.pivot(index='team1', columns='venue', values='win_percentage')

# Plotting the heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(venue_team_performance_pivot, annot=True, cmap="YlGnBu", fmt='.2f')
plt.title('Team Performance at Different Venues', fontsize=16)
plt.show()
```

```python
# Reset the index to get team and venue as columns for easier plotting


# Plotting a grouped bar chart for the win percentage of teams at different venues
plt.figure(figsize=(15, 10))
sns.barplot(x='venue', y='win_percentage', hue='team1', data=venue_team_performance)
plt.title("Win Percentage of Teams at Different Venues", fontsize=16)
plt.xlabel("Venue", fontsize=12)
plt.ylabel("Win Percentage", fontsize=12)
plt.xticks(rotation=90)  # Rotate venue names if needed
plt.legend(title="Teams", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# **Winning Margins Analysis**

```python
# Group by winning team and calculate the mean and median of the result_margin
winning_margin = matches.groupby('winner')['result_margin'].agg(['mean', 'median'])
winning_margin
```

```python
# Define close wins (margin < 10 runs or 2 wickets)
close_wins = matches[matches['result_margin'] < 10]  # This will capture margins under 10
close_wins_by_team = close_wins.groupby('winner').size()
close_wins_by_team
```

```python
# Box plot to visualize winning margins
plt.figure(figsize=(12, 8))
sns.boxplot(x='winner', y='result_margin', data=matches)
plt.title('Winning Margins by Team')
plt.xticks(rotation=90)
plt.show()
```

# **Super Over Analysis**

```python
# Filter matches with super over
super_over_matches = matches[matches['super_over'] == 'Y']
super_over_matches
```

```python
# Count how many times each team has played in a Super Over
super_over_teams = super_over_matches['team1'].value_counts().add (super_over_matches['team2'].value_counts(),fill_value=0)
super_over_teams
```

```python
# Calculate win percentage in Super Overs
super_over_wins = super_over_matches[super_over_matches['winner'].notna()]

# Group by team and calculate win percentage
super_over_win_percentage = super_over_wins.groupby('winner').size() / super_over_teams * 100

super_over_win_percentage
```

```python
# Bar plot to visualize the Super Over performance
plt.figure(figsize=(12, 8))
sns.barplot(x=super_over_win_percentage.index, y=super_over_win_percentage.values)
plt.title('Win Percentage in Super Overs')
plt.ylabel('Win Percentage')
plt.xticks(rotation=90)
plt.show()
```
